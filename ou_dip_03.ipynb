```
# 画像の読み込み
img = cv2.imread("field.jpg")
cv2_imshow(img)
```

```
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)  # HSV色空間へ変換

# 緑色のHSV色空間範囲は(35, 43, 46)から(77, 255, 255)までだから、以下のように閾値範囲を設定
lower = (35, 43, 46)

upper = (77, 255, 255)

# 検出処理
mask =  cv2.inRange(img_hsv, lower, upper)

cv2_imshow(mask)
cv2.imwrite("output.png", mask)

cv2_imshow(mask)
cv2.imwrite("output.png", mask)
```
